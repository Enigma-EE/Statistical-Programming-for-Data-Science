### Exercise 1: Lists
Write code that prompts the user for two lists of integers of equal length, and yields, as output, the element- wise multiplication of the integers.

For example, if the user enters the lists [1,2,3] and [1,2,5], then the output should be 1 4 15, where 1 is the product of 1x1, 4 is the product of 2x2 and 9 is the product of 3x5.

Try to make your code bulletproof, by checking whether the lists are the same length and whether they both contain only integers.
Hint: to check whether they both contain only integers, you can use the code

all(isinstance(item , int) for item in my_list)

where my_list is the list you want to check.

```python
# Prompt the user for two lists of integers
list1 = input("Enter the first list of integers separated by spaces: ")
list2 = input("Enter the second list of integers separated by spaces: ")

# Split the input strings into lists of integers
list1 = [int(num) for num in list1.split()]
list2 = [int(num) for num in list2.split()]

# Check if the two lists have equal length
if len(list1) != len(list2):
  print("Error: The two lists must have equal length")
else:
  # Multiply the integers element-wise
  result = [a * b for a, b in zip(list1, list2)] # zip() returns a list of tuples
  print("The element-wise multiplication of the two lists is:", result)
```

### Exercise 2: Functions for guessing game

```python
import random
import math

def mean(data):
    return sum(data) / len(data) # sum() returns the sum of all elements in the list

def standard_deviation(data):
    m = mean(data)
    variance = sum((x - m) ** 2 for x in data) / len(data) # variance = sum of squared deviations from the mean
    return math.sqrt(variance) # standard deviation = square root of the variance

def z_test(x1, x2, s1, s2, n1, n2):
    numerator = x1 - x2
    denominator = math.sqrt((s1 ** 2 / n1) + (s2 ** 2 / n2))
    return numerator / denominator

def play_game(player_name):
    true_number = random.randint(1, 100)
    guesses = []
    for i in range(3):
        guess = int(input(f"{player_name}, enter your guess: "))
        distance = abs(guess - true_number)
        guesses.append(distance)
        if guess == true_number:
            break
    return guesses

player1_name = input("Enter Player 1's name: ")
player2_name = input("Enter Player 2's name: ")

player1_guesses = play_game(player1_name)
player2_guesses = play_game(player2_name)

if len(player1_guesses) < 3 or len(player2_guesses) < 3: # if one of the players guessed the number in less than 3 guesses
    if len(player1_guesses) < len(player2_guesses): # the player with fewer guesses wins
        winner = player1_name
    elif len(player1_guesses) > len(player2_guesses): # the player with fewer guesses wins
        winner = player2_name
    else:
        winner = "tie" # if both players guessed the number in the same number of guesses, it's a tie
else:
    player1_mean = mean(player1_guesses) # calculate the mean and standard deviation of the guesses
    player2_mean = mean(player2_guesses) # for each player
    player1_sd = standard_deviation(player1_guesses)
    player2_sd = standard_deviation(player2_guesses)

    z_score = z_test(player1_mean, player2_mean, player1_sd, player2_sd, len(player1_guesses), len(player2_guesses))

    if z_score < -1.96:
        winner = player1_name
    elif z_score > 1.96:
        winner = player2_name
    else:
        winner = "tie"

if winner == "tie":
    print("The result is (statistically) a tie.")
else:
    print(f"Congratulations, {winner}! You are the best at guessing.")
```

### Exercise 3: Dictionary – Words count
write a program that will create a dictionary from the words contained in music_review.txt, a text file that contains a music review of the new Madonna album. Each entry in the dictionary should take the form key:value, where key is each unique word in the text file and value is a counter, containing the number of times this word appears.
Hint: instructions on opening, reading & closing files

Your program should:
1. Reduce storage of useless information by not storing the words “a”, “an”, “I”, “the” and any other such words you believe will not be useful.
2. For this exercise, treat words containing punctuation as distinct words, e.g. album and album; would constitute two separate dictionary entries. We can refine this assumption as we progress (or if you’re feeling adventurous, go for it now!).
3. Output to the screen the most commonly used words in the reviews. I will leave it to you to determine appropriate output information and to set the cut-off – you should try to set a meaningful cut-off for this exercise.
Make your program as a function that takes a string and cut-off value, and output most commonly used words.

```python
def word_count(file_name, cutoff):
    # Open the file
    with open(file_name, "r") as f:
        # Read the content
        content = f.read()

    # Split the content into words
    words = content.split()

    # Define a list of words to ignore
    ignore_words = ["a", "an", "I", "the"]

    # Create an empty dictionary to store the word count
    word_dict = {}

    # Iterate through the words
    for word in words:
        # Check if the word is not in the ignore list
        if word not in ignore_words:
            # If the word is already in the dictionary, increment the count
            if word in word_dict:
                word_dict[word] += 1
            # If the word is not in the dictionary, add it with a count of 1
            else:
                word_dict[word] = 1

    # Sort the dictionary by values in descending order
    sorted_word_dict = dict(sorted(word_dict.items(), key=lambda item: item[1], reverse=True))

    # Print the most commonly used words with a count greater than or equal to the cutoff value
    for key, value in sorted_word_dict.items():
        if value >= cutoff:
            print(f"{key}: {value}")

# use the function
word_count("lotto_luck.txt", 3) # 3 is the cutoff value
# cut off value is the minimum number of times a word must appear to be printed
```

```python
def count_words(file_path):
    with open(file_path, encoding='utf-8', errors='ignore') as f:
        text = f.read()

    # Split the text into words and convert to lowercase
    words = text.lower().split()

    # remove special characters
    words = [word.strip('.,?!:;()[]{}') for word in words]

    # Create an empty dictionary to store the word count
    word_dict = {}

    ignore_words = ["a", "an", "I", "the", "and","of","to","that"]
    # Iterate through the words
    for word in words:
        # Check if the word is not in the ignore list 
        if word in ignore_words:
            continue
        # If the word is already in the dictionary, increment the count
        if word in word_dict:
            word_dict[word] += 1
        # If the word is not in the dictionary, add it with a count of 1
        else:
            word_dict[word] = 1

    # Sort the dictionary by values in descending order
    sorted_word_dict = dict(sorted(word_dict.items(), key=lambda item: item[1], reverse=True)) # the lambda function returns the value of the item

    # Return the sorted dictionary
    return sorted_word_dict

# Call the function with the file path as argument
file_path = 'music_review.txt'
sorted_word_dict = count_words(file_path)

# Print the most commonly used words with a count greater than or equal to the cutoff value
for key, value in sorted_word_dict.items():
    if value >= 3:
        print(f"{key}: {value}")
```

## Exercise 4: Pig Latin

```python
def pig_latin(word): 
    vowels = "aeiouAEIOU" # define a string of vowels
    special_cases = ["th", "st", "qu", "pl", "tr"] # define a list of special cases
    word = word.lower()

    if word[0] in vowels:
        return word + "hay" # if the word starts with a vowel, add "hay" to the end
    else:
        for case in special_cases: # check if the word starts with a special case
            if word.startswith(case): # if the word starts with a special case, move the special case to the end and add "ay"
                return word[len(case):] + case + "ay" 
        return word[1:] + word[0] + "ay" # if the word starts with a consonant, move the first letter to the end and add "ay"

def main():
    with open("lotto_luck.txt", "r") as input_file: # open the file in read mode
        text = input_file.read() # read the entire file into a string

    words = text.split() # split the text into a list of words
    pig_latin_words = [pig_latin(word) for word in words] # convert each word to pig latin
    pig_latin_text = " ".join(pig_latin_words) # join the list of pig latin words into a string

    with open("lotto_luck_pig_latin.txt", "w") as output_file:
        output_file.write(pig_latin_text)

if __name__ == "__main__": # run the main() function only if the file is run directly
    main() # if the file is imported, the main() function will not be run
```
